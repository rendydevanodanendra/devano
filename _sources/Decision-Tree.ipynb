# Decision Tree Classifier
Decision Tree adalah salah satu algoritma machine learning berbasis pohon yang digunakan untuk klasifikasi maupun regresi. Dalam konteks klasifikasi, decision tree bekerja dengan memecah data ke dalam cabang-cabang berdasarkan fitur-fitur tertentu, hingga mencapai keputusan pada daun (leaf) yang menunjukkan label kelas.

Visualisasi decision tree menyerupai struktur pohon:
*   Root (akar): node awal, tempat pengambilan keputusan pertama.
*   Internal Nodes: node yang memuat pertanyaan atau kondisi berbasis fitur.
*   Branches (cabang): jalur hasil dari suatu kondisi.
*  Leaf Nodes: hasil akhir berupa label kelas.

###  Langkah 1: Hitung Entropy Awal (Parent Node)
Entropy menggambarkan ketidakteraturan/ketidakmurnian suatu himpunan:

$$
\text{Entropy}(S) = - \sum_{i=1}^{n} p_i \cdot \log_2(p_i)
$$

 Keterangan:

$S$: Sekumpulan data (dataset atau subset).

$n$: Jumlah kelas (label) dalam data.

$p_i$: Proporsi (peluang) dari kelas ke- $i$, yaitu:



  $$
  p_i = \frac{\text{jumlah data dengan kelas ke-}i}{\text{total data}}
  $$


$\log_2$: Logaritma basis 2.



‚ö†Ô∏è Catatan:
Jika $p_i = 0$, maka $p_i \cdot \log_2(p_i)$ dianggap = 0.

Entropy bernilai:


*  Entropy bernilai 0 jika hanya 1 kelas (murni).
*  Maksimum jika distribusi kelas merata (impuritas tertinggi).

##  Langkah 4: Hitung Entropy Masing-Masing Subset Hasil Split

1. **Entropy untuk cabang kiri (misalnya nilai ‚â§ threshold):**

$$
\text{Entropy}_{\text{kiri}} = - \sum_{i=1}^{n} p_{i,\text{kiri}} \cdot \log_2(p_{i,\text{kiri}})
$$

2. **Entropy untuk cabang kanan (misalnya nilai > threshold):**

$$
\text{Entropy}_{\text{kanan}} = - \sum_{i=1}^{n} p_{i,\text{kanan}} \cdot \log_2(p_{i,\text{kanan}})
$$

**Keterangan:**

* $p_{i,\text{kiri}}$ dan $p_{i,\text{kanan}}$ adalah proporsi kelas ke-$i$ pada masing-masing subset
* $n$ adalah jumlah total kelas unik



##  Langkah 5: Hitung Entropy Gabungan Setelah Split

Gabungkan entropy subset dengan bobot jumlah data pada masing-masing subset:

$$
\text{Entropy}_{\text{gabungan}} = \frac{N_{\text{kiri}}}{N} \cdot \text{Entropy}_{\text{kiri}} + \frac{N_{\text{kanan}}}{N} \cdot \text{Entropy}_{\text{kanan}}
$$

**Keterangan:**

* $N_{\text{kiri}}$ dan $N_{\text{kanan}}$: jumlah data pada subset kiri dan kanan
* $N$: jumlah data total



##  Langkah 6: Hitung Information Gain

###  **Rumus Information Gain (IG):**

$$
\text{Information Gain} = \text{Entropy}(S) - \sum_{i=1}^{k} \frac{|S_i|}{|S|} \cdot \text{Entropy}(S_i)
$$



###  **Penjelasan Komponen:**

* **Entropy(S)** = Entropi dari keseluruhan dataset $S$
* $S_i$ = Subset dari data $S$ setelah dibagi berdasarkan suatu atribut
* $|S_i|$ = Jumlah data pada subset ke-$i$
* $|S|$ = Jumlah total data
* $\text{Entropy}(S_i)$ = Entropi subset ke-$i$
* $k$ = Jumlah subset yang terbentuk dari atribut tersebut



Information Gain mengukur **seberapa banyak ketidakpastian (entropy)** yang **berkurang** ketika dataset dibagi berdasarkan atribut tertentu. Atribut dengan **nilai IG tertinggi** dipilih sebagai **akar atau cabang** pada decision tree.




###  Langkah 5: Pilih Split dengan Information Gain Terbesar

* Hitung Information Gain untuk semua kemungkinan fitur dan titik split.
* Pilih split yang memberikan **Information Gain terbesar**.
* Buat node baru berdasarkan split tersebut.



### üîÅ Lanjutkan ke Cabang Berikutnya (Recursion)

* Ulangi proses di atas untuk masing-masing subset (kiri dan kanan) secara rekursif hingga:

  * Semua data dalam subset berada pada kelas yang sama, atau
  * Tidak ada lagi fitur yang bisa di-split



## Perhitungan Entropy dan Information Gain untuk Pemilihan Fitur Pohon Keputusan

````python
import pandas as pd
import numpy as np
from math import log2

# Data asli
data = {
    'X1': [5.00, 6.00, 4.00, 15.00, 14.00, 13.00, 8.00, 9.00, 7.00, 5.00, 14.00, 6.00, 13.00, 8.00, 9.00],
    'X2': [7.00, 9.00, 6.00, 5.00, 6.00, 4.00, 15.00, 16.00, 14.00, 8.00, 5.00, 7.00, 6.00, 13.00, 12.00],
    'Class': [0, 0, 0, 1, 1, 1, 2, 2, 2, 0, 1, 0, 1, 2, 2]
}


df = pd.DataFrame(data)

# Fungsi menghitung entropy
def entropy(class_list):
    values, counts = np.unique(class_list, return_counts=True)
    probs = counts / counts.sum()
    return -np.sum(probs * np.log2(probs))

# Entropy parent (seluruh data)
H_parent = round(entropy(df['Class']), 4)
print("Entropy parent (seluruh data):", H_parent, "\n")

# Fungsi untuk menghitung splits dan information gain
def calculate_splits(df, feature, target='Class'):
    thresholds = sorted(df[feature].unique())
    midpoints = [(thresholds[i] + thresholds[i+1]) / 2 for i in range(len(thresholds) - 1)]

    descriptions = []
    for t in midpoints:
        left = df[df[feature] <= t]
        right = df[df[feature] > t]

        H_left = entropy(left[target])
        H_right = entropy(right[target])

        w_left = len(left) / len(df)
        w_right = len(right) / len(df)

        H_split = w_left * H_left + w_right * H_right
        IG = H_parent - H_split

        desc = f"""[Feature: {feature} | Threshold: {round(t, 2)}]
 - Jumlah Left: {len(left)}
 - Jumlah Right: {len(right)}
 - Entropy Left: {round(H_left, 4)}
 - Entropy Right: {round(H_right, 4)}
 - Weighted Entropy: {round(H_split, 4)}
 - Information Gain: {round(IG, 4)}
--------------------------------------------------"""
        descriptions.append(desc)

    return descriptions

# Hasil dari semua fitur
descriptions_x1 = calculate_splits(df, 'X1')
descriptions_x2 = calculate_splits(df, 'X2')

# Gabungkan dan tampilkan
all_descriptions = descriptions_x1 + descriptions_x2

print("Hasil Perhitungan untuk Semua Fitur dan Threshold:\n")
for desc in all_descriptions:
    print(desc)
````

## üìä Evaluasi Model Decision Tree dengan Self-Evaluation (Akurasi & Laporan Klasifikasi)

````python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Data asli
data = {
    'X1': [5.00, 6.00, 4.00, 15.00, 14.00, 13.00, 8.00, 9.00, 7.00, 5.00, 14.00, 6.00, 13.00, 8.00, 9.00],
    'X2': [7.00, 9.00, 6.00, 5.00, 6.00, 4.00, 15.00, 16.00, 14.00, 8.00, 5.00, 7.00, 6.00, 13.00, 12.00],
    'Class': [0, 0, 0, 1, 1, 1, 2, 2, 2, 0, 1, 0, 1, 2, 2]
}

df = pd.DataFrame(data)

# Pisahkan fitur dan target
X = df[['X1', 'X2']]
y = df['Class']

# Buat dan latih model Decision Tree dengan seluruh data
model = DecisionTreeClassifier(criterion='entropy', random_state=42)
model.fit(X, y)

# Prediksi pada data yang sama (self-evaluation)
y_pred = model.predict(X)

# Evaluasi performa (akurasi dan laporan klasifikasi)
accuracy = accuracy_score(y, y_pred)
print("Akurasi (self-evaluation):", round(accuracy, 4))

print("\nLaporan Klasifikasi:")
print(classification_report(y, y_pred, zero_division=0))
````

## Memvisualisasikan struktur pohon keputusan dalam bentuk diagram pohon dan rule

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text

# Data asli
data = {
    'X1': [5.00, 6.00, 4.00, 15.00, 14.00, 13.00, 8.00, 9.00, 7.00, 5.00, 14.00, 6.00, 13.00, 8.00, 9.00],
    'X2': [7.00, 9.00, 6.00, 5.00, 6.00, 4.00, 15.00, 16.00, 14.00, 8.00, 5.00, 7.00, 6.00, 13.00, 12.00],
    'Class': [0, 0, 0, 1, 1, 1, 2, 2, 2, 0, 1, 0, 1, 2, 2]
}
df = pd.DataFrame(data)

# Fitur dan target
X = df[['X1', 'X2']]
y = df['Class']

# Buat model Decision Tree (pakai seluruh data)
model = DecisionTreeClassifier(criterion='entropy', random_state=42)
model.fit(X, y)

# Visualisasi Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(model,
          feature_names=X.columns,
          class_names=np.unique(y).astype(str),  # <-- Perbaikan di sini
          filled=True,
          rounded=True)
plt.title("Visualisasi Decision Tree (Semua Data)")
plt.show()

# Tampilkan rule dalam bentuk teks
print("\n=== Aturan Pohon Keputusan (Decision Rules) ===")
rules = export_text(model, feature_names=list(X.columns))
print(rules)
````

# Confusion Matrix

````python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Split data menjadi data latih dan uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Buat model dan latih
model = DecisionTreeClassifier(criterion='entropy', random_state=42)
model.fit(X_train, y_train)

# Prediksi data uji
y_pred = model.predict(X_test)

# Buat confusion matrix dan tampilkan
cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot(cmap='Blues')
plt.title(" Confusion Matrix")
plt.show()
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from matplotlib.colors import ListedColormap

# Data asli
data = {
    'X1': [5.00, 6.00, 4.00, 15.00, 14.00, 13.00, 8.00, 9.00, 7.00, 5.00, 14.00, 6.00, 13.00, 8.00, 9.00],
    'X2': [7.00, 9.00, 6.00, 5.00, 6.00, 4.00, 15.00, 16.00, 14.00, 8.00, 5.00, 7.00, 6.00, 13.00, 12.00],
    'Class': [0, 0, 0, 1, 1, 1, 2, 2, 2, 0, 1, 0, 1, 2, 2]
}
df = pd.DataFrame(data)

# Fitur dan target
X = df[['X1', 'X2']]
y = df['Class']

# Buat model Decision Tree (pakai seluruh data)
model = DecisionTreeClassifier(criterion='entropy', random_state=42)
model.fit(X, y)

# Visualisasi pohon keputusan
plt.figure(figsize=(12, 8))
plot_tree(model,
          feature_names=X.columns,
          class_names=[str(cls) for cls in np.unique(y)],
          filled=True,
          rounded=True)
plt.title("Visualisasi Decision Tree (Semua Data)")
plt.show()

# Tampilkan aturan decision tree
print("\n=== Aturan Pohon Keputusan (Decision Rules) ===")
rules = export_text(model, feature_names=list(X.columns))
print(rules)

# Buat grid untuk decision boundary
x_min, x_max = X['X1'].min() - 1, X['X1'].max() + 1
y_min, y_max = X['X2'].min() - 1, X['X2'].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Prediksi nilai klasifikasi untuk setiap titik di grid
grid_points = pd.DataFrame(np.c_[xx.ravel(), yy.ravel()], columns=['X1', 'X2'])
Z = model.predict(grid_points)
Z = Z.reshape(xx.shape)

# Buat colormap
colors = ['#a2d5f2', '#ffcccb', '#d5f2a2']
cmap = ListedColormap(colors[:len(np.unique(y))])

# Plot decision boundary
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=cmap)

# Plot data asli
for class_label in df['Class'].unique():
    subset = df[df['Class'] == class_label]
    plt.scatter(subset['X1'], subset['X2'], label=f'Class {class_label}', s=80)

# Plot data baru (jika ingin)
# test_data = pd.DataFrame({'X1': [7], 'X2': [10]})
# plt.scatter(test_data['X1'], test_data['X2'], c='black', marker='X', s=120, label='Data Baru')

# Label dan grid
plt.title("Decision Boundary + Data Asli")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.grid(True)
plt.show()
````
